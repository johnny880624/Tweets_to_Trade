Create Dataframes With Lag Feature

```python
# Aggregate sentiment scores
daily_sentiment = train_data.groupby(['post_date', 'ticker_symbol'])['sentiment_score'].mean().reset_index()

# Lag feature
daily_sentiment['lag_1'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(1)
daily_sentiment['lag_2'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(2)
daily_sentiment['lag_3'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(3)
daily_sentiment['lag_4'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(4)

# Prepare final DataFrame for model training
train_data_v2 = daily_sentiment.dropna()  # Dropping rows where any lag feature is NaN
train_data_v2.head()
```

```python
# Aggregate sentiment scores
daily_sentiment = test_data.groupby(['post_date', 'ticker_symbol'])['sentiment_score'].mean().reset_index()

# Lag feature
daily_sentiment['lag_1'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(1)
daily_sentiment['lag_2'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(2)
daily_sentiment['lag_3'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(3)
daily_sentiment['lag_4'] = daily_sentiment.groupby('ticker_symbol')['sentiment_score'].shift(4)

# Prepare final DataFrame for model training
test_data_v2 = daily_sentiment.dropna()  # Dropping rows where any lag feature is NaN
test_data_v2.head()
```

Combine with stock price

```python
stock_price.sort_values(by=['Symbol', 'Date'], inplace=True)
train_data_v2['post_date'] = pd.to_datetime(train_data_v2['post_date'])

combined_train_data = pd.merge(left=train_data_v2,
                         right=stock_price,
                         left_on=['post_date', 'ticker_symbol'],
                         right_on=['Date', 'Symbol'],
                         how='left')

combined_train_data.dropna(inplace=True)

# the next day's closing price
combined_train_data['Next_Close'] = combined_train_data.groupby('ticker_symbol')['Close'].shift(-1)
combined_train_data.dropna(subset=['Next_Close'], inplace=True)

combined_train_data.head()
```

```python
stock_price.sort_values(by=['Symbol', 'Date'], inplace=True)
test_data_v2['post_date'] = pd.to_datetime(test_data_v2['post_date'])

combined_test_data = pd.merge(left=test_data_v2,
                         right=stock_price,
                         left_on=['post_date', 'ticker_symbol'],
                         right_on=['Date', 'Symbol'],
                         how='left')

combined_test_data.dropna(inplace=True)

# the next day's closing price
combined_test_data['Next_Close'] = combined_test_data.groupby('ticker_symbol')['Close'].shift(-1)
combined_test_data.dropna(subset=['Next_Close'], inplace=True)

combined_test_data.head()
```

ML

```python
combined_train_data = combined_train_data[combined_train_data['ticker_symbol'] == 'AAPL']
combined_test_data = combined_test_data[combined_test_data['ticker_symbol'] == 'AAPL']
```

```python
combined_train_data['Price_Change'] = combined_train_data['Next_Close'] - combined_train_data['Close']
combined_train_data['Target'] = (combined_train_data['Price_Change'] > 0).astype(int)

combined_test_data['Price_Change'] = combined_test_data['Next_Close'] - combined_test_data['Close']
combined_test_data['Target'] = (combined_test_data['Price_Change'] > 0).astype(int)


combined_train_data = combined_train_data.dropna(subset=['Price_Change'])
combined_test_data = combined_test_data.dropna(subset=['Price_Change'])


y_train = combined_train_data['Target']
y_test = combined_test_data['Target']
```

```python
X_train = combined_train_data[['sentiment_score','lag_1', 'lag_2', 'lag_3', 'lag_4']]
X_test = combined_test_data[['sentiment_score','lag_1', 'lag_2', 'lag_3', 'lag_4']]
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(report)
```

```python
correct_guess =0
for i in range(len(list(y_test))):
  if list(y_test)[i] == y_pred[i]:
    correct_guess +=1
print("total number of prediction: " + str(len(y_pred)))
print("number of corrected prediction: " + str(correct_guess))
print("successful prediction%: " + str(correct_guess/len(y_pred)))
```
